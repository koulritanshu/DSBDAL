```python
# We can also remove outliers using z_score method
from scipy import stats
# creating another copy of original dataset
df = dataset.copy()
df.replace('?',np.nan,inplace=True)
df.dropna(inplace=True)
# this df dataset is not entirely correct
# here na values are just dropped instead they should be replaced by mean values
z_scores = stats.zscore(df['reading score'])
z_scores
```

```python
z_threshold = 3
outliers = df[abs(z_scores) >= z_threshold]

outliers
```

```python
correct_data = df[abs(z_scores) < z_threshold]

correct_data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# firstly applying min-max scaling
scaler = MinMaxScaler()

dataset_minmax = dataset_copy

dataset_minmax['reading score Min-Max'] = scaler.fit_transform(dataset_minmax['reading score'].values.reshape(-1,1))

dataset_minmax
```

```python
dataset_minmax['writing score Min-Max'] = scaler.fit_transform(dataset_minmax['writing score'].values.reshape(-1,1))
dataset_minmax['math score Min-Max'] = scaler.fit_transform(dataset_minmax['math score'].values.reshape(-1,1))
dataset_minmax
```

```python
scaler = StandardScaler()

dataset_std = dataset_copy

dataset_std['reading standard'] = scaler.fit_transform(dataset_std['reading score'].values.reshape(-1,1))

dataset_std['writing standard'] = scaler.fit_transform(dataset_std['writing score'].values.reshape(-1,1))

dataset_std['math standard'] = scaler.fit_transform(dataset_std['math score'].values.reshape(-1,1))

dataset_std
```
